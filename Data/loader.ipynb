```python
import numpy as np
import pandas as pd
```

```python
lines = []
with open("C:\\DP-Data\\nipt\\mapping\\grch38\\stats-wgs\\samples\\9153\\genome_results.txt") as f:
    lines = f.readlines()
count = 0
for line in lines:
    count += 1
```

```python
labels = ['ids', 'bases', 'contigs', 'windows', 'reads',
'mapped_reads', 'mapped_perc', 'secondary_alignments', 'mapped_paired_reads_first', 'mapped_paired_reads_second', 'mapped_paired_reads_both',
'mapped_paired_reads_singletons', 'mapped_bases', 'sequenced_bases', 'aligned_bases', 'duplicated_reads_flagged', 'mean_insert_size', 'std_insert_size',
'median_insert_size', 'mean_mapping_quality', 'A_num', 'A_perc', 'C_num', 'C_perc', 'T_num', 'T_perc', 'G_num', 'G_perc', 'N_num', 'N_perc', 'general_error_rate',
'mismatches', 'insertions', 'mapped_reads_insertion_perc', 'deletions', 'mapped_reads_deletion_perc', 'homopolymer_indels', 'mean_coverageData', 'std_coverageData']

data = dict()
for label in labels:
  data[label] = []
```

```python
contig_coverage = dict()

for x in range(22):
  chr_num = 'chr' + str(x+1)
  sm_dict = dict()
  sm_dict['len'] = []
  sm_dict['mapped'] = []
  sm_dict['coverage'] = []
  sm_dict['deviation'] = []
  contig_coverage[chr_num] = sm_dict

additional_tags = ['X','Y','M']

for tag in additional_tags:
  chr_num = 'chr' + tag
  sm_dict = dict()
  sm_dict['len'] = []
  sm_dict['mapped'] = []
  sm_dict['coverage'] = []
  sm_dict['deviation'] = []
  contig_coverage[chr_num] = sm_dict
```

```python
data['ids'].append('9153')

data['bases'].append(int(''.join(lines[11].split()[-2].split(','))))
data['contigs'].append(int(lines[12].split()[-1]))

data['windows'].append(int(lines[17].split()[-1]))

data['reads'].append(int(''.join(lines[19].split()[-1].split(','))))
data['mapped_reads'].append(int(''.join(lines[20].split()[-2].split(','))))
data['mapped_perc'].append(data['mapped_reads'][-1]/data['reads'][-1])
data['secondary_alignments'].append(int(lines[21].split()[-1]))

data['mapped_paired_reads_first'].append(int(''.join(lines[23].split()[-1].split(','))))
data['mapped_paired_reads_second'].append(int(''.join(lines[24].split()[-1].split(','))))
data['mapped_paired_reads_both'].append(int(''.join(lines[25].split()[-1].split(','))))
data['mapped_paired_reads_singletons'].append(int(''.join(lines[26].split()[-1].split(','))))

data['mapped_bases'].append(int(''.join(lines[28].split()[-2].split(','))))
data['sequenced_bases'].append(int(''.join(lines[29].split()[-2].split(','))))
data['aligned_bases'].append(int(''.join(lines[30].split()[-2].split(','))))
data['duplicated_reads_flagged'].append(int(''.join(lines[31].split()[-1].split(','))))

data['mean_insert_size'].append(float(''.join(lines[36].split()[-1].split(','))))
data['std_insert_size'].append(float(''.join(lines[37].split()[-1].split(','))))
data['median_insert_size'].append(float(''.join(lines[38].split()[-1].split(','))))

data['mean_mapping_quality'].append(float(''.join(lines[43].split()[-1].split(','))))

data['A_num'].append(int(''.join(lines[48].split()[-3].split(','))))
data['A_perc'].append(float(lines[48].split()[-1].strip('(.%)'))/100)
data['C_num'].append(int(''.join(lines[49].split()[-3].split(','))))
data['C_perc'].append(float(lines[49].split()[-1].strip('(.%)'))/100)
data['T_num'].append(int(''.join(lines[50].split()[-3].split(','))))
data['T_perc'].append(float(lines[50].split()[-1].strip('(.%)'))/100)
data['G_num'].append(int(''.join(lines[51].split()[-3].split(','))))
data['G_perc'].append(float(lines[51].split()[-1].strip('(.%)'))/100)
data['N_num'].append(int(''.join(lines[52].split()[-3].split(','))))
data['N_perc'].append(float(lines[52].split()[-1].strip('(.%)'))/100)

data['general_error_rate'].append(float(lines[59].split()[-1]))
data['mismatches'].append(int(''.join(lines[60].split()[-1].split(','))))
data['insertions'].append(int(''.join(lines[61].split()[-1].split(','))))
data['mapped_reads_insertion_perc'].append(float(lines[62].split()[-1].strip('(.%)'))/100)
data['deletions'].append(int(''.join(lines[63].split()[-1].split(','))))
data['mapped_reads_deletion_perc'].append(float(lines[64].split()[-1].strip('(.%)'))/100)
data['homopolymer_indels'].append(float(lines[65].split()[-1].strip('(.%)'))/100)

data['mean_coverageData'].append(lines[70].split()[-1])
data['std_coverageData'].append(lines[71].split()[-1])

contigs_coverage_index = lines.index('>>>>>>> Coverage per contig\n')
contig_counter = 2

for key in contig_coverage.keys():
  contig_coverage[key]['len'].append(int(lines[contigs_coverage_index + contig_counter].split()[1]))
  contig_coverage[key]['mapped'].append(int(lines[contigs_coverage_index + contig_counter].split()[2]))
  contig_coverage[key]['coverage'].append(float(lines[contigs_coverage_index + contig_counter].split()[3]))
  contig_coverage[key]['deviation'].append(float(lines[contigs_coverage_index + contig_counter].split()[4]))
  contig_counter += 1
```
