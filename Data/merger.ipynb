```python
import pandas as pd
import numpy as np
```

```python
filename = "metadata.csv"
metadata_df = pd.read_csv(filename,index_col=0, encoding = "ISO-8859-1",  sep=';')
metadata_df.head(2)
```

```python
metadata_df.columns
```

```python
important_metadata = metadata_df[['sample_id','core-result',
       'core-t13', 'core-t18', 'core-t21']]
```

```python
filename = "data.csv"
data_df = pd.read_csv(filename,index_col=0, encoding = "ISO-8859-1",  sep=',')
data_df.head(2)
```

```python
data_df.rename(columns = {'ids':'sample_id'}, inplace = True)
```

```python
data_df.head(2)
```

```python
merged= pd.merge(data_df, metadata_df, on='sample_id', how='left')
```

```python
merged.describe()
```

```python
merged['core-result'].isna().sum()
```

```python
merged = merged[merged['core-result'].notna()]
```

```python
merged = merged[merged['core-result'] != 'uninformative']
merged = merged[merged['core-t13'] != 'uninformative']
merged = merged[merged['core-t18'] != 'uninformative']
merged = merged[merged['core-t21'] != 'uninformative']
```

```python
merged['core-result'].value_counts()
```

```python
merged.to_csv('merged.csv')  
```
